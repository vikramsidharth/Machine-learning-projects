# Utilization of Machine Learning for Smart Food Spoilage Detection System leveraging AMTH-IoT Data

```python
#Importing libraries
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
dataset=pd.read_csv("dataset.csv")
```

```python
dataset=dataset.drop('Unnamed: 0',axis=1)
```

```python
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
# Create an instance of LabelEncoder
label_encoder = LabelEncoder()

# Define the columns to be encoded
columns_to_encode = ['TimeOfDay', 'FoodType', 'Spoiled']

# Apply label encoding to each column
for column in columns_to_encode:
    dataset[column] = label_encoder.fit_transform(dataset[column])

# Display the encoded DataFrame
dataset
```

```python
#defining x and y
```

```python
X=dataset.drop("Spoiled",axis=1)
y=dataset['Spoiled']
```

```python
#Datasplitting
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
# Standardize the feature values (optional but recommended for SVMs)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```
